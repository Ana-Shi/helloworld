```python
def lr_model(regressor_data):
    regressor_data = pd.get_dummies(regressor_data)
    Y = regressor_data['cost']
    X = regressor_data.drop(['cost'],axis=1)

    X_train = X[1:409]
    X_test = X[410:]
    Y_train = Y[1:409]
    Y_test = Y[410:]

    from sklearn.linear_model import LinearRegression

    lr_model = LinearRegression()
    lr_model.fit(X_train,Y_train)
    y_predict = lr_model.predict(X_test)

    
    return y_predict,Y_test
```

```python
from sklearn.metrics import r2_score
def show_result(y_predict,Y_test):
    r2_lr  = r2_score(Y_test,y_predict)
    from sklearn.metrics import mean_absolute_error,mean_squared_error
    print('MAE:{}'.format(mean_absolute_error(y_predict,Y_test)))
    print('MSE:{}'.format(mean_squared_error(y_predict,Y_test)))
    plt.figure(figsize=(5,5))
    plt.plot(range(0,len(y_predict)),y_predict,c='b',marker='o',label='predict')
    plt.plot(range(0,len(y_predict)),Y_test,c='y',marker='o',label='real')
    plt.legend(loc=[1,0])
    plt.show()
```
