```python
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
import numpy as np
X = np.linspace(0,10,100)
Y = np.sin(X)
cosy = np.cos(X)
```

```python
plt.plot(X,Y)
plt.plot(X,cosy,color='red',linestyle=':')
plt.show()
```

```python
# 调整坐标系大小
plt.plot(X,Y)
plt.plot(X,cosy,color='red',linestyle=':')
# plt.xlim(-5,15)
# plt.ylim(0,1.5)
plt.axis([-1,11,-2,2])
plt.show()
```

```python
#x,y代表
# 调整坐标系大小
plt.plot(X,Y)
plt.plot(X,cosy,color='red',linestyle=':')
# plt.xlim(-5,15)
# plt.ylim(0,1.5)
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')
plt.ylabel('y value')
plt.show()
```

```python
# 曲线加标注

plt.plot(X,Y,label = 'sin(x)')
plt.plot(X,cosy,color='red',linestyle=':',label='cos x')
# plt.xlim(-5,15)
# plt.ylim(0,1.5)
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()#加legend
plt.show()
```

```python
# 加标题

plt.plot(X,Y,label = 'sin(x)')
plt.plot(X,cosy,color='red',linestyle=':',label='cos x')
# plt.xlim(-5,15)
# plt.ylim(0,1.5)
plt.axis([-1,11,-2,2])
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()#加legend
plt.title('wecome to the ML World')
plt.show()
```

# 读取数据并进行可视化探索

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris.keys()
```

```python
print(iris.DESCR)
```

```python
X = iris.data[:,:2]
```

```python
iris.data.shape
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
y = iris.target
```

```python
plt.scatter(X[y==0,0],X[y==0,1],c='red')
plt.scatter(X[y==1,0],X[y==1,1],c='blue')
plt.scatter(X[y==2,0],X[y==2,1],c='green')
plt.show()
```

```python
# 设置散点图样式
plt.scatter(X[y==0,0],X[y==0,1],c='red',marker = 'o')
plt.scatter(X[y==1,0],X[y==1,1],c='blue',marker = 'x')
plt.scatter(X[y==2,0],X[y==2,1],c='green',marker = '+')
plt.show()
```
